## Disease-Specific Survival (DSS) in TCGA Data

- **DSS**: Disease-Specific Survival
  - **0**: Patient is alive or died from a non-disease-related cause
  - **1**: Patient died due to the disease under study


```python
df['DSS']
```

```python
df['DSS'].isna().sum()
```

```python
# drop records where DSS is null
df = df.dropna(subset=['DSS'])
```

```python
df['DSS'].isna().sum()
```

```python
df.head()
```

```python
# Calculate total and percent missing values
total_missing = df.isnull().sum().sort_values(ascending=False)
percent_missing = (df.isnull().sum() / df.isnull().count()).sort_values(ascending=False)

# Combine into a DataFrame
missing_data_df = pd.concat([total_missing, percent_missing], axis=1, keys=["Total Missing", "Percent Missing"])

# Filter to only include columns with missing values greater than 0
missing_data_df = missing_data_df[missing_data_df["Total Missing"] > 0]

# Display the resulting DataFrame
missing_data_df
```

```python
filtered_missing_data_df = missing_data_df[missing_data_df['Total Missing'] < 25]
filtered_missing_data_df.head(20)
```

```python
# Get the columns with missing data
columns_to_drop = missing_data_df.index.tolist()

# Drop these columns from the original dataframe
df = df.drop(columns=columns_to_drop)
```

```python
cols_to_label_encode = [
    'icd_o_3_histology',
    'icd_o_3_site',
    'neoplasm_histologic_grade',
    'other_dx',
    'pathologic_M',
    'pathologic_N',
    'pathologic_T'
]
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to each column
for col in cols_to_label_encode:
    df[col] = le.fit_transform(df[col])
```

```python
df['pathologic_N'].unique()
```

```python
cols_to_onehot = [
    'gender',
    'tissue_prospective_collection_indicator',
    'vital_status'
]
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Initialize the OneHotEncoder
onehot_encoder = OneHotEncoder(sparse_output=False, drop='first') # drop='first' to avoid multicollinearity, optional

# Fit and transform the data
encoded_features = onehot_encoder.fit_transform(df[cols_to_onehot])

# Create a DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_features, columns=onehot_encoder.get_feature_names_out(cols_to_onehot))

encoded_df
```

```python
# Drop the original columns and concatenate the encoded features
df = df.drop(columns=cols_to_onehot).reset_index(drop=True)
df_encoded = pd.concat([df, encoded_df], axis=1)

df_encoded.head()
```
